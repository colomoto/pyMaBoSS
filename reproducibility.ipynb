```python
from maboss.gsparser import build_network
reprod_sim = build_network('test/reprod_all')
# This loads the file given in supplementary matterial (S1 File)
# in http://journals.plos.org/ploscompbiol/article?id=10.1371/journal.pcbi.1004571
```

```python
reprod_sim.network.set_output(['EMT', 'Metastasis', 'Invasion', 'Apoptosis'])
```

```python
wt_results = reprod_sim.run()
```

```python
wt_results.plot_piechart()
```

```python
reprod_sim.mutate('p53', 'OFF')
```

```python
p53_LoF = reprod_sim.run()
```

```python
p53_LoF.plot_piechart()
```

```python
reprod_sim.mutate('p53', 'WT')
reprod_sim.network.set_istate('p53', [1, 0]) #p53 is not forced on or off but its initial state is forced to be off
reprod_sim.mutate('NICD', 'ON')
```

```python
nicd_GoF = reprod_sim.run()
```

```python
nicd_GoF.plot_piechart()
```

```python
reprod_sim.mutate('p53', 'OFF')
p53_LoF_nicd_GoF = reprod_sim.run()
```

```python
p53_LoF_nicd_GoF.plot_piechart()
```
