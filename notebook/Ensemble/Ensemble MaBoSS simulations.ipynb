# Ensemble MaBoSS simulations

```python
import maboss, time
```

## Simulating the ensemble

```python
t0 = time.time()

simulation = maboss.Ensemble(
    path="Invasion",
    sample_count=100000, 
    thread_count=6,
    max_time=50, time_tick=1,
    discrete_time=1,
    outputs=['Apoptosis', 'CellCycleArrest', 'Invasion'],
    random_sampling=True
)

result = simulation.run()

print("Simulation executed in %.2fs" % (time.time() - t0))
```

```python
result.plot_piechart()
```

```python
result.plot_trajectory()
```

## Simulating the ensemble, keeping individual results to be analyzed

```python
t0 = time.time()

simulation = maboss.Ensemble(
    path="Invasion",
    individual_results=True,
)

simulation.param.update({
    'sample_count': 100000,
    'thread_count': 6,
    'max_time': 50, 
    'time_tick': 1,
})

simulation.set_outputs(['Apoptosis', 'CellCycleArrest', 'Invasion'])
result = simulation.run()

print("Simulation executed in %.2fs" % (time.time() - t0))
```

```python
result.plotSteadyStatesDistribution(figsize=(20, 10))
```

```python
result.plotSteadyStatesNodesDistribution(figsize=(20, 10))
```
