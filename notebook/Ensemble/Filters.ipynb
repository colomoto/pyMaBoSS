```python
import maboss
```

### We simulate the Notch++ p53-- mutation for the invasion model

```python
simulation = maboss.Ensemble(
    path="Invasion",
    use_physrandgen=0, 
    sample_count=100000,
    thread_count=6,
    istate={
        'DNAdamage': (0.5, 0.5),
        'ECMicroenv': (0.5, 0.5),
    },
    max_time=50, time_tick=1,
    outputs=[
        'Apoptosis', 'CellCycleArrest', 'Invasion'
    ],
    mutations={"NICD": "ON", "p53": "OFF"},
    individual_results=True,
    random_sampling=True
)

result = simulation.run()
```

```python
result.plot_piechart()
```

```python
result.getSteadyStatesNodesDistribution(filter="Invasion < 1.0")
```

### We filter the models of the ensemble to get the one who don't have 100% invasion

```python
result.createSubEnsemble("FilteredInvasion", node_filter="Invasion < 1.0")
```

```python
simulation_sub = maboss.Ensemble(
    path="FilteredInvasion",
    use_physrandgen=0, 
    sample_count=100000,
    thread_count=6,
    istate={
        'DNAdamage': (0.5, 0.5),
        'ECMicroenv': (0.5, 0.5),
    },
    max_time=50, time_tick=1,
    outputs=[
        'Apoptosis', 'CellCycleArrest', 'Invasion'
    ],
    mutations={"NICD": "ON", "p53": "OFF"},
    individual_results=True,
    random_sampling=True
)

result_sub = simulation_sub.run()
```

```python
result_sub.plot_piechart()
```
